```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import os

# ========== 1. 读取合并后的数据集 ==========
data_path = r"C:\Users\33759\Desktop\btc数据挖掘\修改后数据\BTC_FNG_合并数据.xlsx"
merge_df = pd.read_excel(data_path)

# ========== 2. 数据清洗：重复值、缺失值处理 ==========
# （1）重复值处理：按日期去重
merge_df = merge_df.drop_duplicates(subset='日期', keep='first')
print(f"去重后数据量：{len(merge_df)} 条")

# （2）缺失值处理：删除含缺失值的行（或填充）
# 方法：删除缺失值（保证数据完整性）
merge_df = merge_df.dropna(subset=['情绪值', '涨跌幅（百分比）', '次日涨跌幅（百分比）'])
print(f"删除缺失值后数据量：{len(merge_df)} 条")


# ========== 3. 数据清洗：归一化处理（可选，用于消除量纲差异） ==========
# 对“情绪值”和“次日涨跌幅（百分比）”进行归一化（缩放到0-1区间）
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# ========== 5. 保存清洗后的数据集 ==========
clean_save_path = os.path.join(r"C:\Users\33759\Desktop\btc数据挖掘\修改后数据", "BTC_FNG_清洗后数据.xlsx")
merge_df.to_excel(clean_save_path, index=False, engine='openpyxl')
print(f"\n✅ 清洗后数据已保存至：{clean_save_path}")
```
